```python
import numpy as np
import pandas as pd
import regex
import matplotlib.pyplot as plt
```

```python
f = open('temp/EmbeddingIntrinsicResults', 'r')
results = pd.DataFrame(columns = ['d', 'c', 'w', 'test', 'rho', 'missing', 'total'])
for line in f:
    d, c ,w = map(int, regex.findall(r'([\d]+)d_([\d]+)c_([\d]+)w', line)[0])
    next(f)
    temp = []
    for i in range(3):
        temp.append(next(f).split())
    for i in temp:
        results = results.append({'d': d,
                                 'c': c,
                                  'w': w,
                                  'test': i[0],
                                  'rho': float(i[1]),
                                  'missing': int(i[2]),
                                  'total': int(i[3])         
                                 }
                                 ,ignore_index = True)
```
